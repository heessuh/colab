```
counts = dict() #create an empty dictionary
print('Enter a line of text:')
line = input('') #Assign whatever string the user types in to the variable line

words = line.split() #split the string into words

print('Words:', words)

print('Counting...')
for word in words:
  counts[word] = counts.get(word,0) + 1 #get the word and its occurence and assign it to the variable counts
print('Frequency of words', counts)
```

```
import operator
sorted_counts = sorted(counts.items(), key=operator.itemgetter(1)) #sort them in ascending order
print(sorted_counts)
```
