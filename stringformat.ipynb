```
#upper() converts the entire string to upper case
x = "hello"
x.upper()
```

```
"hello".upper()
```

```
#capitalize() capitalizes only the beginning of string
"hello".capitalize()
```

```
#lower() converts the entire string to lower case
"HELLO".lower()
```

```
#len() gives the length of the string  REMEMBER len() goes BEFORE the string
len("123456789  ")
```
